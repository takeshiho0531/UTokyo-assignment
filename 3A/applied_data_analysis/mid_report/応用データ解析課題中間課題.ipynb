### (1)

```python
medv_data = df['MEDV']

# 期待値（平均値）
mean_medv = np.mean(medv_data)

# 中央値
median_medv = np.median(medv_data)

# 標本分散
sample_variance_medv = np.var(medv_data, ddof=0)

# 不偏分散
unbiased_variance_medv = np.var(medv_data, ddof=1)

# 歪度
skewness_medv = medv_data.skew()

# 尖度
kurtosis_medv = medv_data.kurtosis()


print("期待値（平均値）:", mean_medv)
print("中央値:", median_medv)
print("標本分散:", sample_variance_medv)
print("不偏分散:", unbiased_variance_medv)
print("歪度:", skewness_medv)
print("尖度:", kurtosis_medv)
```

```python
from scipy.stats import norm

plt.figure(figsize=(10, 5))
plt.hist(medv_data, bins=30, density=True, alpha=0.7, color='blue', label='MEDV Distribution')

# 正規分布の確率密度関数を計算して描画
mu, sigma = np.mean(medv_data), np.std(medv_data)
x = np.linspace(min(medv_data), max(medv_data), 100)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, 'r--', label='Normal Distribution')

plt.xlabel('MEDV (Housing Price)')
plt.ylabel('Probability Density')
plt.title('MEDV Distribution vs. Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```
